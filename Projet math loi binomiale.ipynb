```python
import numpy as np
import math
import matplotlib.pyplot as plt

def simulate_binomial(n, p, num_simulations):
    results = np.random.binomial(n, p, num_simulations)
    return results

n = 20 # nombre d'essais
p = 0.3 # probabilité de sinistre pour un essai individuel
num_simulations = 1000000 # nombre de simulations souhaitées

simulations = simulate_binomial(n, p, num_simulations)
print("Simulations: ", simulations)
print("Esperance: ", n*p)
print("Variance: ", n*p*(1-p))
print("Ecart type: ", math.sqrt(n*p*(1-p)))


x = np.arange(0, n + 1, 1)

# Calculer la densité empirique
histogram = np.histogram(simulations, bins=x, density=True)

# Calculer la densité théorique
def binomial_pdf(x, n, p):
    return math.comb(n, x) * p**x * (1-p)**(n-x)
y = [binomial_pdf(i, n, p) for i in x]

# Dessiner la densité
plt.plot(x, y, 'r')
plt.bar(histogram[1][:-1], histogram[0], width=1)
plt.xlabel("Nombre de sinistres")
plt.ylabel("Densité de probabilité")
plt.title("Densité théorique vs empérique de la loi binomiale")
plt.show()
```
