```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Paramètres de la loi de Weibull
shape = 2.5
scale = 100
n=10000

# Générer des données selon la loi de Weibull
data = np.random.weibull(shape, size=n) * scale
print(data)

# Tracer la densité empirique
plt.hist(data, bins=50, density=True, alpha=0.5, label='Densité empirique')

# Tracer la densité théorique
x = np.linspace(0, np.max(data), num=1000)
pdf = weibull_min.pdf(x/scale, shape)
plt.plot(x, pdf, label='Densité théorique')

plt.legend()
plt.show()
```

```python
# Estimation des quantiles extrêmes
p = 0.01
q1 = np.percentile(data, p*100)
p = 0.99
q2 = np.percentile(data, p*100)

# Calcul des quantiles théoriques
q1_theorique = scale * weibull_min.ppf(q=0.01, c=shape, loc=0, scale=scale)
q2_theorique = scale * weibull_min.ppf(q=0.99, c=shape, loc=0, scale=scale)

# Affichage des résultats
print(f"Quantile 1% (estimé) : {q1}")
print(f"Quantile 1% (théorique) : {q1_theorique}")

print(f"Quantile 99% (estimé) : {q2}")
print(f"Quantile 99% (théorique) : {q2_theorique}")
```
